```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from selenium.webdriver.common.action_chains import ActionChains
from selenium.webdriver.common.keys import Keys
from selenium.common.exceptions import NoSuchElementException, TimeoutException
import time
```

```python
driver = webdriver.Chrome()
```

```python
url = 'https://quotes.toscrape.com/login'
```

```python
driver.get(url)
time.sleep(1)
email_field = driver.find_element(By.ID, 'username')
password_field = driver.find_element(By.ID, 'password')

email_field.send_keys('temp@gmail.com')
password_field.send_keys('123')

password_field.send_keys(Keys.ENTER)
authors = []
descriptions = []
tags_of_quotes = []
good_reads = []
about = []

while True:
    try:
        next_page_element = WebDriverWait(driver, 3).until(
            EC.presence_of_element_located((By.CSS_SELECTOR, '.next a'))
        )
        next_page_url = next_page_element.get_attribute('href')
        
        quotes = driver.find_elements(By.CLASS_NAME, 'quote')

        for quote in quotes:
            description = quote.find_element(By.CLASS_NAME, 'text').text
            author = quote.find_element(By.CLASS_NAME, 'author').text
            all_a_from_spans = quote.find_elements(By.CSS_SELECTOR, 'span a')
            about_of_author = all_a_from_spans[0].get_attribute('href')
            good_to_read = all_a_from_spans[1].get_attribute('href')
            tags = quote.find_elements(By.CLASS_NAME, 'tag')
            for tag in tags:
                tags_of_quotes.append(tag.get_attribute('href'))
            
            
            good_reads.append(good_to_read)
            about.append(about_of_author)
            descriptions.append(description)
            authors.append(author)
        
        driver.get(next_page_url)
    except TimeoutException:
        print("Next page element not found. No more pages to navigate.")
        break
    
print(authors, descriptions, tags_of_quotes, good_reads, about)
```
